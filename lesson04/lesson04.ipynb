4. Рост взрослого населения города X имеет нормальное распределение. 
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
а) больше 182 см;
б) больше 190 см;
в) от 166 см до 190 см;
г) от 166 см до 182 см;
д) от 158 см до 190 см;
е) не выше 150 см или не ниже 190 см;
ё) не выше 150 см или не ниже 198 см;
ж) ниже 166 см.

```python
# Z = (X - M)/S
M = 174
S = 8

# а) P(рост больше 182 см) = 1 - P(рост меньше, либо равен 182 см)
Z1 = (182 - M)/S
print('z =', Z1)
# по таблице находим значение 0.8413 - вероятность, что рост человека не более 182 см.
print('а) P =', 1 - 0.8413)

# б) P(рост больше 190 см) = 1 - P(рост меньше, либо равен 190 см)
Z2 = (190 - M)/S
print('z =', Z2)
# по таблице находим значение 0.9772 - вероятность, что рост человека не более 190 см.
print('б) P =', 1 - 0.9772)

# в) P(рост от 166 см до 190 см) = P(рост меньше, либо равен 190 см) - P(рост меньше, либо равен 166 см)
Z3 = (166 - M)/S
print('z =', Z3)
# Φ(-x) = 1 - Φ(x)
# по таблице находим значение 1 - 0.8413 - вероятность, что рост человека не более 166 см.
print('в) P =', 0.9772 - (1 - 0.8413))

# г) P(рост от 166 см до 182 см) = P(рост меньше, либо равен 182 см) - P(рост меньше, либо равен 166 см)
# а) и в)
print('г) P =', 0.8413 - (1 - 0.8413))

# д) P(рост от 158 см до 190 см) = P(рост меньше, либо равен 190 см) - P(рост меньше, либо равен 158 см)
Z5 = (158 - M)/S
print('z =', Z5)
print('д) P =', 0.9772 - (1 - 0.9772))

# е) P(рост не выше 150 см или не ниже 190 см) = P(рост меньше 150 см) + б)
Z6 = (150 - M)/S
print('z =', Z6)
# по таблице находим значение 1 - 0.9987 - вероятность, что рост человека не более 150 см.
print('е) P =', (1 - 0.9987) + (1 - 0.9772))

# ё) P(рост не выше 150 см или не ниже 198 см) = P(рост меньше 150 см) + (1 - P(рост меньше 198 см))
Z7 = (198 - M)/S
print('z =', Z7)
print('ё) P =', (1 - 0.9987) + (1 - 0.9987))

# ж) P(рост ниже 166 см) = a), то есть P(рост ниже 166 см) = P(рост больше 182 см)
print('ж) P =', 1 - 0.8413)
```
